## A) A rigidez do pórtico (k_t);

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as signal 
from scipy import *
import sympy as sy
import plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Características do corpo

m = 0.462     # [kg]
w_n = 24.5 # [rad/s]    | 3.9     #[hz] | 24.5 #[rad/s]
t = np.linspace(0, 11, 1000)

#Impulso -

f_chap = 0.1 #[N*s]



```

```python
# w_n = np.sqrt(k_t/m)

k_t = (w_n**2)*m

print('Rigidez do pórtico (k_t) é : {} [N/m]'.format(k_t))
```

## B) O amortecimento crítico (c_c);

```python
# c_c = 2*np.sqrt(k_t*m)

c_c = 2*np.sqrt(k_t*m)

print('O amortecimento crítico (c_c) é : {} [N*s/m]'.format(c_c))
```

## C) As condições iniciais :
            x_0; e

```python
# v_0 = f_chap/m

v_0 = f_chap/m
x_ponto_0 = v_0

print('As condições iniciais: v_0  é : {} [m/s]'.format(v_0))
```

## D) Plote as respostas x(t) para o pórtico considerando :
            zeta_1 = 1%;
            zeta_2 = 5%;
            zeta_3 = 10%;e
            zeta_4 = 20%.

```python
#x(t) = (f_chap/(m*w_n*np.sqrt(1-zeta**2)))*np.sin(w_n*np.sqrt(1-zeta**2)*t)

zeta_1 = 0.01
zeta_2 = 0.05
zeta_3 = 0.10
zeta_4 = 0.20


#x = ((v_0/(w_n*np.sqrt(1-zeta**2)))*np.exp(-zeta*w_n*t)*np.sin(w_n*np.sqrt(1-zeta**2)*t)
```

```python
x=[]
for zeta in [zeta_1,zeta_2,zeta_3,zeta_4]:

    x.append((v_0/(w_n*np.sqrt(1-zeta**2)))*np.exp(-zeta*w_n*t)*np.sin(w_n*np.sqrt(1-zeta**2)*t))
```

```python
# figura 6: signal veloc filtrado 4 - 200 Hz

fig = make_subplots(rows=4, cols=1)


fig.add_trace(go.Scatter(x=t,y=x[0], mode="lines", name="1%", line_width=1, marker_color="#511CFB"), row=1, col=1)
fig.add_trace(go.Scatter(x=t,y=x[1], mode="lines", name="5%", line_width=1, marker_color="#AF0038"), row=2, col=1)
fig.add_trace(go.Scatter(x=t,y=x[2], mode="lines", name="10%", line_width=1, marker_color="#00A08B"), row=3, col=1)
fig.add_trace(go.Scatter(x=t,y=x[3], mode="lines", name="20%", line_width=1, marker_color="#00A08B"), row=4, col=1)

fig.update_layout(title_text="Respostas x(t) para variações do fator de amortecimento",
                  title_font_size=30, height=700,
                  xaxis_title="Tempo (s)",
                  yaxis_title="Amplitude [m/s]")

# Definir o título do eixo Y do subplot na linha 2, coluna 1
fig.update_yaxes(title="Amplitude [m/s]", row=2, col=1)
fig.update_xaxes(title="Tempo (s)", row=2, col=1)

# Definir o título do eixo Y do subplot na linha 2, coluna 1
fig.update_yaxes(title="Amplitude [m/s]", row=3, col=1)
fig.update_xaxes(title="Tempo (s)", row=3, col=1)

fig.update_yaxes(title="Amplitude [m/s]", row=4, col=1)
fig.update_xaxes(title="Tempo (s)", row=4, col=1)
fig.show()
```

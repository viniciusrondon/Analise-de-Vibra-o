```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import *
import scipy.io
from scipy.fftpack import fft, rfft, ifft, fftfreq, fftshift
from scipy.signal import detrend
from datetime import datetime
import csv
import xlrd
import pandas as pd
import re
```

```python
# medição

sample = 48000
Delta = 0.000488
T = sample * Delta
fs = 1.000 / Delta # sample rate, Hz



```

```python
def open_measure(file_name):

    df = pd.read_csv(file_name,sep=';')
    
    return df

def butter_bandpass(cutoff_low, cutoff_high, fs, order=4):
    nyq = 0.5 * fs
    low = cutoff_low / nyq
    high = cutoff_high / nyq
    b, a = butter(order, [low, high], btype='bandpass', analog=False)
    return b, a

def butter_bandpass_filter(data, low, high, fs, order=5):
    b, a = butter_bandpass(low, high, fs, order=order)
    #y = lfilter(b, a, data)
    y = filtfilt(b, a, data)
    return y

#def parse_float(value):
       # """Parse float from a value."""
        #return float(re.sub(r",", ".", value))

```

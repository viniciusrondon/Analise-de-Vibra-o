```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
tn = 0.21 #segundos
```

```python
('O período natural é {} segundos' ).format(tn)
```

```python
wn = (2*np.pi)/tn #frequencia natural
('A frequencia natural é {} rad/s' ).format(wn)
```

```python
fn = 1/((2*np.pi)/wn)
('A frequencia é {} Hz' ).format(fn)
```

```python
wn1 = np.sqrt(1.5)*wn #frequencia natural com acrescimo de 50% da constante de mola
('A frequencia natural com acrescimo de 50% da constante de mola é {} rad/s' ).format(wn1)
```

```python
tn1 = (2*np.pi)/wn1
('O período natural com acrescimo de 50% da constante de mola é {} segundos' ).format(tn1)
```

```python
fn1 = 1/tn1
('A frequencia com acrescimo de 50% da constante de mola é {} Hz' ).format(fn1)
```

```python
wn2 = np.sqrt(0.5)*wn #frequencia natural com acrescimo de 50% da constante de mola
('A frequencia natural com decrescimo de 50% da constante de mola é {} rad/s' ).format(wn2)
```

```python
tn2 = (2*np.pi)/wn2
('O período natural com decrescimo de 50% da constante de mola é {} segundos' ).format(tn2)
```

```python
fn2 = 1/tn2
('A frequencia com decrescimo de 50% da constante de mola é {} Hz' ).format(fn2)
```

```python
wn3 = np.sqrt(1/0.5)*wn #frequencia natural com acrescimo de 50% na massa
('A frequencia natural com decrescimo de 50% na massa é {} rad/s' ).format(wn3)
```

```python
tn3 = (2*np.pi)/wn3
('O período natural com decrescimo de 50% massa é {} segundos' ).format(tn3)
```

```python
fn3 = 1/tn3
('A frequencia com decrescimo de 50% na massa é {} Hz' ).format(fn3)
```

```python
wn4 = np.sqrt(1/1.5)*wn #frequencia natural com acrescimo de 50% da constante de mola
('A frequencia natural com acrescimo de 50% na massa é {} rad/s' ).format(wn4)
```

```python
tn4 = (2*np.pi)/wn4
('O período natural com acrescimo de 50% na massa é {} segundos' ).format(tn4)
```

```python
fn4 = 1/tn4
('A frequencia com acrescimo de 50% massa é {} Hz' ).format(fn4)
```

```python
array = np.arange(-62,62,2) #porcentagem de aumento
wn5 = ((2*np.pi)/0.21) * (np.sqrt(1+(array/100))) #alteracao de constante elastica
wn6 = ((2*np.pi/0.21)*(np.sqrt(1/(1+array/100)))) #alteracao de massa
```

```python
plt.plot(array,wn5)
plt.xlabel("aumento de %")
plt.ylabel("frequencia natural rad/s")
plt.xticks(np.arange(-50, 60, step=10))
plt.grid(True)
plt.show()
```

```python
plt.plot(array,wn6)
plt.xlabel("aumento de %")
plt.ylabel("frequencia natural rad/s")
plt.xticks(np.arange(-50, 60, step=10))
plt.grid(True)
plt.show()
```
